```python
import os
import h5py
import numpy as np
from PIL import Image

for entry in os.listdir(inputPath):
    if os.path.isfile(os.path.join(inputPath, entry)):
        print(entry)
        
        fileId = entry[:-4]
        imageId = fileId + '.png'
        print(fileId, imageId)
        
        imageOut = outputPath + fileId + '/images/'
        maskOut = outputPath + fileId + '/masks/'
        
        os.makedirs(imageOut)
        os.makedirs(maskOut)
        
        file = inputPath + entry
        hdf = h5py.File(file,'r')
  
        array = hdf["cjdata/image"][:]
        maxi = np.amax(array)
        array = array * 255.0 / maxi
        img = Image.fromarray(array.astype('uint8'), 'L')
        
        imgURI = imageOut + imageId
        img.save(imgURI, "PNG")
        
        array = hdf["cjdata/tumorMask"][:]
        maxi = np.amax(array)
        array = array * 255.0 / maxi
        img = Image.fromarray(array.astype('uint8'), 'L')
        
        maskURI = maskOut + imageId
        img.save(maskURI, "PNG")
```
